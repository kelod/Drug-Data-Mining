```python
import pandas as pd

PATH = ""
df = pd.read_csv(PATH)
df
```

## Modellek kipróbálása módosítások nélkül

```python
df.columns
```

```python
input_attributes = ['interaction_group_score', 'TARGET_CLASS', 'ACT_VALUE', 'drug_approved', 'drug_avr_mass', 'drug_predicted_logp',
                   'drug_rule_of_five', 'drug_nr_targets', 'drug_nr_enzymes', 'drug_nr_carriers', 'drug_nr_transporters', 'POLAR_SURFACE_AREA',
       'ROTATABLE_BOND_COUNT', 'HYDROGEN_BOND_DONORS',
       'HYDROGEN_BOND_ACCEPTORS', 'RING_COUNT', 'ATOM_COUNT',
       'CHIRAL_CENTER_COUNT']

target_label=['interaction_types']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[input_attributes], df[target_label], test_size=0.2, random_state=42)
```

```python
X_train = pd.get_dummies(X_train, columns=['TARGET_CLASS'])
X_test = pd.get_dummies(X_test, columns=['TARGET_CLASS'])
```

One-hot encoding a categorical típusú változóra, mert később a modellek nem tudnának string értékekkel dolgozni.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

log_reg = LogisticRegression(random_state=42)
log_reg.fit(X_train, y_train)

svm = make_pipeline(StandardScaler(), SVC(gamma='auto'))
svm.fit(X_train, y_train)

dec_tree = DecisionTreeClassifier(random_state=42)
dec_tree.fit(X_train, y_train)

knn = KNeighborsClassifier(n_neighbors=5) # Legyen 5 szavazat - páratlan kell
knn.fit(X_train, y_train)

# Betanítás
```

```python
from sklearn.metrics import accuracy_score

print("Logistic regression accuracy: ", accuracy_score(y_test, log_reg.predict(X_test)))
print("Support Vector Machine accuracy: ", accuracy_score(y_test, svm.predict(X_test)))
print("Decision tree accuracy: ", accuracy_score(y_test, dec_tree.predict(X_test)))
print("K Nearest Neighbor accuracy: ", accuracy_score(y_test, knn.predict(X_test)))
```

Egyelőre csak simán random szétválasztott train-test settel tehát a fent látható accuracyk születtek. Legjobban az SVM és a Dec. Tree teljesített, a logistic regression a legszerényebben.
Jó hír, hogy egy teljesen triviális és használhatatlan modelnél is (amelyik pl mindegyikre agonist-ot jósol) jobban teljesítenek ilyen százalékokkal.

```python
for i in range(0,24):
    print(X_train.columns[i], log_reg.coef_[3][i])
```

Az látható, hogy a log.reg melyik változókhoz, milyen súlyokat rendelt. Kiemelkedik pl. az ACT_VALUE, valamint a gyógyszer targetjeinek száma, illetve enzimeinek száma, a log.reg úgy találta, hogy ezek a tulajdonságok nyomnak a legtöbbet a latba, míg például a hidrogén donorok száma, és a gyógyszer tömege szinte elhanyagolható (3 nagyságrendbeli különbség)

## ROC görbe

ROC görbét általában binary classification esetén rajzolnak, multilabel esetén a one vs all metodikát használva lehet több ROC görbét kirajzolni.
A rajzolás rengeteg időt vett igénybe jupyter notebookban (sőt, egyszer sikerült, aztán a többi próbánál hosszú várakozás után abbahagytam), ezért most kifejezetten a google colab használatát javasolnám (annak nagyjából 5-6 percre volt szüksége) a kirajzolásra.
A görbén egyébként látható, hogy még így is tanult a model azért valamit SVM esetén (az átlótól egész jól eltér), és az AUC érték sem olyan rossz.

Ez a cella egyszerre egy labelle vonatkozó görbét rajzol ki, a kommenttel jelölt helyen (alul) lehet az indexeléssel (0-3) változtatni, hogy melyik labelre vonatkozó görbét szeretnénk ha kirajzolná.

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import cycle

from sklearn import svm, datasets
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from scipy import interp
from sklearn.metrics import roc_auc_score

dfc = df.copy()
from sklearn.preprocessing import OrdinalEncoder
import itertool

ord_enc = OrdinalEncoder()
encoded = ord_enc.fit_transform(dfc[['interaction_types']])
flatted = list(itertools.chain(*encoded))

input_attributes2 = ['interaction_group_score', 'ACT_VALUE', 'drug_avr_mass', 'drug_predicted_logp', 'drug_nr_targets', 'drug_nr_enzymes', 'drug_nr_carriers', 'drug_nr_transporters', 'POLAR_SURFACE_AREA',
       'ROTATABLE_BOND_COUNT', 'HYDROGEN_BOND_DONORS',
       'HYDROGEN_BOND_ACCEPTORS', 'RING_COUNT', 'ATOM_COUNT',
       'CHIRAL_CENTER_COUNT']


X = df[input_attributes2].to_numpy()
y = flatted

# Binarize the output
y = label_binarize(y, classes=[0, 1, 2, 3])
n_classes = y.shape[1]


# shuffle and split training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5,
                                                    random_state=0)

# Learn to predict each class against the other
classifier = OneVsRestClassifier(svm.SVC(kernel='linear', probability=True,
                                 random_state=42))
y_score = classifier.fit(X_train, y_train).decision_function(X_test)

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), y_score.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

plt.figure()
lw = 3
plt.plot(fpr[1], tpr[1], color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[1]) # Itt lehet mind a három index kicserélésével változtatni!
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```
